```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Prayagraj_Traffic_Data.csv')
df.sample(25)
```

```python
df.isnull().sum()
```

```python
# Function to generate a unique random ID
def generate_unique_random_id(existing_ids, low=1, high=10000):
    while True:
        random_id = np.random.randint(low, high)
        if random_id not in existing_ids: 
            return random_id

# Assuming 'df' is your dfFrame
existing_ids = set(df['record_id'].dropna())  # Get existing Record IDs without missing values
df['record_id'] = df['record_id'].apply(
    lambda x: generate_unique_random_id(existing_ids) if pd.isnull(x) else x
)

# Convert 'timestamp' column to DateTime format and drop rows with missing timestamps
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df.dropna(subset=['timestamp'], inplace=True)
```

```python
# Mapping between intersection names and IDs
area_to_id_map = {
    'Allahabad Junction': 6, 'Naini Bridge': 5, 'Chowk': 3, 'Tagore Town': 4,
    'Katra': 2, 'Phaphamau': 7, 'Minto Park': 10, 'MNNIT Gate': 9, 
    'Civil Lines': 1, 'Jhunsi': 8
}

# Fill missing 'intersection_id' using 'intersection_name' mapping
df['intersection_id'] = df['intersection_name'].map(area_to_id_map).fillna(df['intersection_id'])

# Fill missing 'intersection_name' using reverse mapping from 'intersection_id'
df['intersection_name'] = df['intersection_id'].map({v: k for k, v in area_to_id_map.items()}).fillna(df['intersection_name'])

# Resulting df Frame
print(df)
```

```python
# Latitude Mapping
area_to_latitude_map = {
    'Allahabad Junction': 25.4358, 'Naini Bridge': 25.4088, 'Chowk': 25.4314, 'Tagore Town': 25.4675,
    'Katra': 25.4550, 'Phaphamau': 25.4945, 'Minto Park': 25.4352, 'MNNIT Gate': 25.4920, 
    'Civil Lines': 25.4486, 'Jhunsi': 25.4326
}

df['latitude'] = df['intersection_name'].map(area_to_latitude_map).fillna(df['latitude'])

# Longitude Mapping
area_to_longitude_map = {
    'Allahabad Junction': 81.8307, 'Naini Bridge': 81.8592, 'Chowk': 81.8437, 'Tagore Town': 81.8867,
    'Katra': 81.8463, 'Phaphamau': 81.8686, 'Minto Park': 81.8820, 'MNNIT Gate': 81.8639, 
    'Civil Lines': 81.8333, 'Jhunsi': 81.9061
}

df['longitude'] = df['intersection_name'].map(area_to_longitude_map).fillna(df['longitude'])
```

```python
# Step 1: Calculate mean values for each intersection_name
mean_values = df.groupby('intersection_name')[['vehicle_count', 'pollution_level', 'car_count', 'bus_count', 'truck_count', 'motorcycle_count']].mean()

# Step 2: Fill missing values for the specified columns using mean values
for col in ['vehicle_count', 'pollution_level', 'car_count', 'bus_count', 'truck_count', 'motorcycle_count']:
    df[col] = df.apply(
        lambda row: mean_values.loc[row['intersection_name'], col] if pd.isnull(row[col]) else row[col], axis=1
    )

# Step 3: Fill missing values in 'average_speed' with the overall mean
df['average_speed'] = df['average_speed'].fillna(df['average_speed'].mean())
```

```python
# Replace missing values in 'signal_cycle_time' with a constant value
df['signal_cycle_time'] = df['signal_cycle_time'].fillna(120.0)

# List of columns to fill with their mean
columns_to_fill = ['green_time', 'yellow_time', 'red_time']

# Replace missing values in each of these columns with their respective mean
for col in columns_to_fill:
    df[col] = df[col].fillna(df[col].mean())
```

```python
# Replace missing values in 'weather_condition' with its mode
weather_mode = df['weather_condition'].mode()[0]  # Find the mode (most frequent value)
df['weather_condition'] = df['weather_condition'].fillna(weather_mode)
```

```python
#  Define the mapping for days of the week
days_mapping = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}

#  Replace the values in the 'day_of_week' column directly
df['day_of_week'] = df['day_of_week'].map(days_mapping)
```

```python
#  Define the mapping dictionary
event_mapping = {
    np.nan: 0,  # Handle NaN explicitly
    'New Year Celebration': 1,
    'Roadwork': 2,
    'Local Market Fair': 3,
    'Accident': 4
}

#  Replace the values in the 'event' column directly
df['event'] = df['event'].map(event_mapping).fillna(0).astype(int)
```

```python
# Extract day of the week from the 'timestamp' column
df['day_of_week'] = df['timestamp'].dt.day_name()

# Assign 'is_holiday' based on whether the day is Sunday
df['is_holiday'] = df['day_of_week'].map({"Sunday": 1}).fillna(0)

# Drop rows with any remaining missing values
df.dropna(inplace=True)
```

```python
# Create the 'is_weekend' column based on the 'day_of_week' column
df['is_weekend'] = df['day_of_week'].map({'Saturday': 1, 'Sunday': 1}).fillna(0).astype(int)

# Extract the hour from the 'timestamp' column and create a new column 'hour_of_day'
df['hour_of_day'] = df['timestamp'].dt.hour
```

```python
df.isnull().sum()
```

```python
columns_to_convert = [
    'record_id', 'intersection_id', 'signal_cycle_time', 
    'green_time', 'yellow_time', 'red_time', 'car_count', 'bus_count', 
    'truck_count', 'motorcycle_count', 'event', 'is_holiday', 'is_weekend', 
    'hour_of_day'
]

for col in columns_to_convert:
    df[col] = df[col].astype(int)
```

```python
# Save the modified DataFrame to a new CSV file
df.to_csv('Prayagraj_Traffic_Data_Cleaned.csv', index=False)

print("All changes have been saved to 'Prayagraj_Traffic_Data_Cleaned.csv'")
```
